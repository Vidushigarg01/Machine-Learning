```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
df
```

Data preprocessing : Fill NAN values with median

```python
df.isnull().sum()
```

```python
median = df["bedrooms"].median()
```

```python
df["bedrooms"] = df["bedrooms"].fillna(median)
```

```python
df.isnull().sum()
```

```python
X = df[["area","bedrooms","age"]]
Y = df.price
```

```python
model = linear_model.LinearRegression()
model.fit(X,Y)
```

```python
area = int(input("Enter area : "))
bedrooms = int(input("Enter bedrooms : "))
age = int(input("Enter age : "))
model.predict([[area,bedrooms,age]])
```

price = slope1*feature1+slope2*feature2+slope3*feature3+intercept

```python
slope1 = model.coef_[0]
slope2 = model.coef_[1]
slope3 = model.coef_[2]
intercept = model.intercept_
```

```python
price = slope1*area+slope2*bedrooms+slope3*age+intercept
price
```
